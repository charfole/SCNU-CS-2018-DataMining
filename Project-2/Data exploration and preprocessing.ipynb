```
import pandas as pd
import xlrd
import seaborn as sns

from datetime import date,datetime

file = 'data_depression.xlsx'

def read_excel(colNumber):

	wb = xlrd.open_workbook(filename=file)#打开文件

	sheet1 = wb.sheet_by_index(0)#通过索引获取表格

	cols = sheet1.col_values(colNumber)[3:]#获取列内容

	return cols

# 读第5列
col5 = read_excel(5)
# 读第6列
col6 = read_excel(6)
# 读第7列
col7 = read_excel(7)
# 读第8列
col8 = read_excel(8)


# 读第9列
col9 = read_excel(9)
# 读第10列
col10 = read_excel(10)
# 读第11列
col11 = read_excel(11)
# 读第12列
col12 = read_excel(12)

print(col12)
```

```
all_list = []
all_list.append(col5)
all_list.append(col6)
all_list.append(col7)
all_list.append(col8)
all_list.append(col9)
all_list.append(col10)
all_list.append(col11)
all_list.append(col12)
df = pd.DataFrame(all_list)
df = df.T
df.columns = ['δ, µV','θ, µV','α, µV','β(LF), µV','δ','θ','α','β(LF)']
print(df)
```

```
 
import pandas as pd
import pylab as plt
 
c5 = pd.Series(col5) #利用Series将列表转换成新的、pandas可处理的数据
c9 = pd.Series(col9)
 
corr_gust = round(c5.corr(c9), 4) #计算标准差，round(a, 4)是保留a的前四位小数
 
print('corr_gust :', corr_gust)
 
#最后画一下两列表散点图，直观感受下，结合相关系数揣摩揣摩
plt.scatter(col5, col9)
plt.title('δ,µV and δ corr_gust :' + str(corr_gust)) #给图写上title
plt.show()
```
